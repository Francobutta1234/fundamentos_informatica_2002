```python
lista1 = [3, 7, 12, 15, 21]
lista2 = [1, 4, 10, 14, 19]

def sumar(l1, l2):
    resultado = []
    for i in range(len(l1)):
        resultado.append(l1[i] + l2[i])

    return resultado
sumar(lista1, lista2)       
```

```python
lista1 = [3, 7, 12, 15, 21]
lista2 = [1, 4, 10, 14, 19]

def resta(l1, l2):
    resultado = []
    for i in range(len(l1)):
        resultado.append(l1[i] - l2[i])

    return resultado
resta(lista1, lista2)
```

```python
lista1 = [3, 7, 12, 15, 21]
lista2 = [1, 4, 10, 14, 19]

def multiplicar(l1, l2):
    resultado = []
    for i in range(len(l1)):
        resultado.append(l1[i] * l2[i])

    return resultado
multiplicar(lista1, lista2)
```

```python
lista1 = [3, 7, 12, 15, 21]
lista2 = [1, 4, 10, 14, 19]

def dividir(l1, l2):
    resultado = []
    for i in range(len(l1)):
        resultado.append(round(l1[i] / l2[i]))

    return resultado
dividir(lista1, lista2)
```

```python
#otra manera de resolver

def series(lista1, lista2, operacion):
    ser1 = pd.Series(lista1)
    ser2 = pd.Series(lista2)
    if operacion == "multiplicar":
        return ser1 * ser2
    elif operacion == "dividir":
        return ser1 / ser2
    elif operacion == "sumar":
        return ser1 + ser2
    elif operacion == "restar":
        return ser1 - ser2
series([3, 7, 12, 15, 21], [1, 4, 10, 14, 19], "sumar")
```

```python
def comparacion(lista1, lista2):
    ser1 = pd.Series(lista1)
    ser2 = pd.Series(lista2)
    return "ser1 > ser2: ", ser1 > ser2, ";ser2 > ser1: ", ser2 > ser1, ";ser1 = ser2: ", ser1 == ser2
print(comparacion([3, 7, 9, 14, 25], [1, 7, 10, 16, 19]))                    
```
